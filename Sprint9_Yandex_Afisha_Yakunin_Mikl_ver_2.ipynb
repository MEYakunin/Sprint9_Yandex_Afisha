### 2.1 *Конвертация валют и создание признака revenue_rub*

```python
# Загрузка данных
exchange_rate_tenge = pd.read_csv('https://code.s3.yandex.net/datasets/final_tickets_tenge_df.csv')
```

```python
# Выводим первые 5 строк датафрейма
exchange_rate_tenge.head(5)
```

```python
# Преобразуем датафрейм с курсами валют в Series с индексацией по датам
exchange_rate_tenge = exchange_rate_tenge.set_index('data')['curs']
exchange_rate_tenge.index = pd.to_datetime(exchange_rate_tenge.index)
```

```python
# Создаем функцию для конвертации выручки в рубли
def calculate_revenue_rub(row):
    if row['currency_code'] == 'kzt':
        exchange_rate = exchange_rate_tenge.get(row['order_dt'])
        return round(row['revenue'] / 100 * exchange_rate, 2)
    else:
        return round(row['revenue'], 2)
```

```python
# создание нового столбца с сконвертированной выручкой
df['revenue_rub'] = df.apply(calculate_revenue_rub, axis=1)
```

```python
# Выводим случайные 5 строчек датафрейма
df.sample(5)
```

```python
# Формируем новый датафрейм с агрегированными признаками
df_user_profile = df_quantile_99.groupby('user_id').agg(
    first_order_dt = ('order_dt', 'min'),
    last_order_dt = ('order_dt', 'max'),
    first_device = ('device_type_canonical', 'first'),
    first_region = ('region_name', 'first'),
    first_service_name = ('service_name', 'first'),
    first_event = ('event_type_main', 'first'),
    total_order = ('order_id', 'count'),
    avg_revenue = ('revenue_rub', 'mean'),
    avg_tickets = ('tickets_count', 'mean'),
    avg_days_between = ('days_since_prev', 'mean')
).assign(
    is_two = lambda x: x['total_order'] >= 2,
    is_five = lambda x: x['total_order'] >=5
)
```

```python
def segment_count_plot(segment, df):
    top_10 = df[segment].value_counts().head(10).index
    
    plt.figure(figsize=(12, 7))
    ax = sns.countplot(
            data=df,
            x=segment,
            hue=segment,
            order=top_10
        )
    
    # Проверяем доступность bar_label
    if hasattr(ax, 'bar_label'):
        # Используем современный метод
        for container in ax.containers:
            absolute_values = [count.get_height() for count in container]
            percentages = [val / len(df) * 100 for val in absolute_values]
            labels = [f'{abs_val:,.0f} ({pct:.1f}%)' 
                     for abs_val, pct in zip(absolute_values, percentages)]
            ax.bar_label(container, labels=labels, padding=3)
    else:
        # Используем альтернативный метод для старых версий
        for container in ax.containers:
            absolute_values = [count.get_height() for count in container]
            percentages = [val / len(df) * 100 for val in absolute_values]
            
            for i, (abs_val, pct) in enumerate(zip(absolute_values, percentages)):
                label = f'{abs_val:,.0f} ({pct:.1f}%)'
                x = container.get_x() + container.get_width() / 2
                y = abs_val
                ax.text(x, y + 0.5, label, ha='center', va='bottom', fontsize=9)
```

```python
def segment_precent_plot(segment, df):
    plt.figure(figsize=(18, 7))
    
    # Вычисляем общее количество в каждом сегменте и вычисляем процент от сегмента
    counts = df.groupby([segment, 'is_two']).size().reset_index(name='count')
    total_per_segment = counts.groupby(segment)['count'].transform('sum')
    counts['percent'] = (counts['count'] / total_per_segment) * 100
    
    # Оставляем только топ-10 сегментов
    top_segments = df[segment].value_counts().head(10).index
    counts = counts[counts[segment].isin(top_segments)]
    
    # Строим столбчатую диаграмму
    ax = sns.barplot(
        data=counts,
        x=segment,
        y='percent',
        hue='is_two',
        order=top_segments
    )
    
    # Добавляем подписи
    for container in ax.containers:
        ax.bar_label(container, fmt='%.1f%%', label_type='edge', padding=3)
```

### 4.4 *Корреляционный анализ количества покупок и признаков пользователя*

```python
# Формируем корреляционную матрицы
interval_cols = ['total_order', 'avg_revenue', 'firstavg_tickets_service_name', 'avg_days_between']
df_corr = df_user_profile.drop(columns=['first_order_dt', 'last_order_dt', 'avg_tickets', 'is_two', 'is_five', 'segment_total_order']).phik_matrix(interval_cols=interval_cols)
```

```python
# Cтроим тепловую карту
df_corr_heatmap = df_corr.loc[df_corr.index != 'total_order'][['total_order']].sort_values(by=['total_order'], ascending=False)
sns.heatmap(df_corr_heatmap,
           annot=True,
           fmt='.2f',
           cmap='coolwarm',
           linewidths=0.5,
           cbar=False)

plt.title('Тепловая карта коэффициента phi_k для данных total_order')
plt.xlabel('Рейтинг категорий')

plt.show()
```
